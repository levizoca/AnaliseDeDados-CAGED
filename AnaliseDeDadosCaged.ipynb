```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
low_memory=False
%matplotlib notebook
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

```python
filename = r'C:\CAGEDMOV-12-2020.txt'
df = pd.read_csv(filename, delimiter=';', error_bad_lines=False)
```

```python
df.columns
```

```python
for par in enumerate(df.columns): print (par)
```

```python
troca_regiao = {1:'Norte', 2:'Nordeste', 3:'Sudeste',
              4:'Sul', 5:'Centro-Oeste', 9:'Não Identificado'}
```

```python
df.loc[(df.região.isnull()), 'região'] = 9
df['região'] = df['região'].apply(lambda x: troca_regiao[x])
```

```python
df.groupby('região').size().sort_values().plot(kind='barh', figsize=(10,5))
```

```python
df.groupby('raçacor').size().sort_values().tail(10).plot(kind='bar', figsize=(10,5))
```
